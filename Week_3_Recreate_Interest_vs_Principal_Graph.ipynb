```python
import pandas as pd
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSdgsW4J8l9V8UlKLuwdYdqyemwkAuTHzS4s6plb-BQIJ8DhI8BkNDvO_Ei3M4lBJElVEaueyOg1Jkj/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
x = df['Month']
y1 = df['3% Cumulative Interest']
y2 = df['5% Cumulative Interest']
fig, ax = plt.subplots()
ax.plot(x,y1,label='30 Year 3% APR')
ax.plot(x,y2,label='30 Year 5% APR')
ax.legend()
ax.set_xlabel('Month')
ax.set_ylabel('Dollars $')
ax.set_title('Cumulative Interest on 400k$ Loan');
```
