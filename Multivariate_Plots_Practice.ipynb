```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fpath='/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/data-cleaning-core.csv'
df =pd.read_csv(fpath)
df.head()
```

```python
g = sns.catplot(data=df,x='Sex',kind='count', hue='Survived')
g.ax.set_title('Survival Distribution by Sex');
```

which group had a greater number of survivors (Survived=1)?
- female

```python
g = sns.catplot(data=df,kind='bar',y='Parents or Children' ,x='Survived',hue='Survived')
g.ax.set_title('Parents or children Survival');
```

which group had a higher average number of Parents or Children?
- Survivors

```python
corr = df.corr(numeric_only=True)
fig, ax = plt.subplots()
ax = sns.heatmap(corr, annot=True, cmap='coolwarm')
ax.set_title('Correlation Heatmap');
```

which column had the highest positive correlation to the "Survived" column?
