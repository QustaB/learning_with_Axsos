```python
import pandas as pd
file_path = '/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/insurance_mod.csv'
df=pd.read_csv(file_path)
df.head()
```

### Average Age

```python
average_age = df['age'].mean()
print(average_age)
```

### Oldest Person

```python
oldest_person = df['age'].max()
print(oldest_person)
```

### Youngest Person

```python
youngest_person = df['age'].min()
print(youngest_person)
```

### Average Charge

```python
average_charge = df['charges'].mean()
print(average_charge)
```

### Average Charge Grouped by Sex

```python
average_charge_by_sex = df.groupby('sex')['charges'].mean()
print(average_charge_by_sex)
```

### Average Charge Grouped by Smokers

```python
average_charge_by_smokers = df.groupby('smoker')['charges'].mean()
print(average_charge_by_smokers)
```

### Average Charge Grouped by Smokers And Sex

```python
average_charge_by_smokers_and_sex = df.groupby(['smoker', 'sex'])['charges'].mean()
print(average_charge_by_smokers_and_sex)
```
