```python
import pandas as pd
import seaborn as sns
```

```python
fpath = '/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/data-cleaning-core.csv'
df = pd.read_csv(fpath)
df.head()
```

## Plot the counts for the number of passengers who Survived vs. those that died

```python
val_counts = df['Survived'].value_counts()
val_counts
```

```python
ax = val_counts.plot(kind='bar')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_title('# of Passenger Survived Vs. Died')
```

```python
ax = sns.countplot(data=df, x='Survived')
ax.set_title('# of Passenger Survived Vs. Died');
```

## Plot a histogram of the distribution of Age

```python
ax = df['Age'].hist(bins = 'auto', edgecolor = 'black')
ax.ticklabel_format(style='plain')
ax.set_title('Distribution of Age')
ax.set_xlabel('Age')
ax.set_ylabel('Count');
```

```python
ax = sns.histplot(data=df, x='Age')
#ax.ticklabel_format(style='plain')
ax.set_title('Distribution of Age');
```

## Plot a boxplot for the Fare column

```python
ax = df['Fare'].plot(kind='box')
ax.set_title('Fare Boxplot')
ax.set_ylabel('Dollars ($)');
```

```python
ax = df.boxplot(column='Fare')
ax.set_title('Fare Boxplot')
ax.set_ylabel('Dollars ($)');
```

```python
ax = sns.boxplot(data=df, y='Fare')
ax.set_title('Fare Boxplot')
ax.set_ylabel('Dollars ($)')
ax.set_xlabel('Fare');
```
